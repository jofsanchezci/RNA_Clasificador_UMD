```python
#Motor de Inferencia
#Tomas las reglas y las procesa para tomar una decisión

def inferir(reglas, hechos):
  return reglas[hechos]

reglas = {
    'fiebre': 'tomar_acetaminofem',
    'tos': 'tomar_jarabe'}

#hechos = {'fiebre': True, 'tos': True}
hechos = 'tos'

accion = inferir(reglas, hechos)
print(accion)

```

```python
class SistemaExperto:
    def __init__(self):
        # Base de conocimientos: reglas
        self.reglas = {
            'fiebre': 'tomar_acetaminofem',
            'tos': 'tomar_jarabe',
            'dolor_de_cabeza': 'tomar_analgesico',
            'dolor_de_estomago': 'tomar_antiacido',
            'fatiga': 'descansar'
        }

        # Base de conocimientos: hechos iniciales
        self.hechos = {
            'fiebre': False,
            'tos': False,
            'dolor_de_cabeza': False,
            'dolor_de_estomago': False,
            'fatiga': False
        }

    def agregar_hecho(self, hecho, valor):
        self.hechos[hecho] = valor

    def inferir(self):
        hechos_a_procesar = list(self.hechos.items())
        while hechos_a_procesar:
            hecho, valor = hechos_a_procesar.pop(0)
            if valor:
                self.aplicar_regla(hecho)

    def aplicar_regla(self, hecho):
        if hecho in self.reglas:
            accion = self.reglas[hecho]
            print(f'Hecho: {hecho} -> Acción: {accion}')
            self.hechos[accion] = True

# Crear el sistema experto
sistema_experto = SistemaExperto()

# Agregar hechos
sistema_experto.agregar_hecho('fiebre', True)
sistema_experto.agregar_hecho('dolor_de_cabeza', True)
sistema_experto.agregar_hecho('tos', True)

# Inferir conclusiones
sistema_experto.inferir()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido
G = nx.DiGraph()

# Agregar nodos (conceptos)
G.add_node("Animal")
G.add_node("Mamífero")
G.add_node("Pájaro")
G.add_node("Perro")
G.add_node("Gato")
G.add_node("Canario")
G.add_node("Pez")

# Agregar aristas (relaciones)
G.add_edge("Perro", "Mamífero", relation="es un")
G.add_edge("Gato", "Mamífero", relation="es un")
G.add_edge("Canario", "Pájaro", relation="es un")
G.add_edge("Mamífero", "Animal", relation="es un")
G.add_edge("Pájaro", "Animal", relation="es un")
G.add_edge("Pez", "Animal", relation="es un")

# Agregar características
G.add_edge("Perro", "Ladra", relation="hace")
G.add_edge("Gato", "Maulla", relation="hace")
G.add_edge("Canario", "Canta", relation="hace")
G.add_edge("Pájaro", "Vuela", relation="puede")

# Función para buscar rutas entre dos nodos
def buscar_rutas(origen, destino):
    try:
        rutas = list(nx.all_simple_paths(G, source=origen, target=destino))
        if not rutas:
            return f"No hay rutas entre {origen} y {destino}."
        return rutas
    except nx.NetworkXNoPath:
        return f"No hay rutas entre {origen} y {destino}."

# Función para encontrar subcategorías de un nodo
def subcategorias(nodo):
    descendientes = list(nx.descendants(G, nodo))
    if not descendientes:
        return f"{nodo} no tiene subcategorías."
    return descendientes

# Ejemplo de búsqueda de rutas
origen = "Ladra"
destino = "Vuela"
print(f"Rutas entre {origen} y {destino}: {buscar_rutas(origen, destino)}")

# Ejemplo de razonamiento para encontrar subcategorías
nodo = "Animal"
print(f"Subcategorías de {nodo}: {subcategorias(nodo)}")

# Dibujar el grafo
pos = nx.spring_layout(G)
plt.figure(figsize=(10, 8))

# Dibujar nodos y aristas
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=3000, font_size=10, font_weight='bold', edge_color='gray')

# Dibujar etiquetas de las aristas
edge_labels = nx.get_edge_attributes(G, 'relation')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

plt.title("Red Semántica de Animales")
plt.show()
```

**Funciones de Búsqueda y Razonamiento:**

* **buscar_rutas(origen, destino):** encuentra todas las rutas simples entre dos nodos en la red semántica. Utiliza la función nx.all_simple_paths de networkx para encontrar todas las rutas posibles.
* **subcategorias(nodo):** encuentra todas las subcategorías de un nodo dado. Utiliza la función nx.descendants de networkx para encontrar todos los nodos descendientes.

# Controladores Difusos

Un controlador difuso, o controlador de lógica difusa, es un sistema que utiliza lógica difusa en lugar de lógica booleana para tomar decisiones. La lógica difusa permite trabajar con valores que pueden estar parcialmente verdaderos y parcialmente falsos al mismo tiempo, lo cual es muy útil para manejar la incertidumbre y la imprecisión en sistemas complejos.

## Conceptos Clave

1. **Lógica Difusa**: Extiende la lógica booleana para manejar la verdad parcial. En lugar de solo 0 y 1 (falso y verdadero), se utilizan valores entre 0 y 1 para representar grados de verdad.
2. **Conjuntos Difusos**: Grupos de elementos con un grado de pertenencia. Por ejemplo, la temperatura "alta" podría tener un valor de pertenencia de 0.7 para 30°C, indicando que es bastante alta, pero no completamente.
3. **Funciones de Pertenencia**: Definen cómo cada punto en el espacio de entrada se asigna a un grado de pertenencia entre 0 y 1. Estas funciones pueden tener formas variadas (triangulares, trapezoidales, gaussianas, etc.).

## Estructura de un Controlador Difuso

Un controlador difuso típicamente consta de las siguientes partes:

1. **Fuzzificación**: Convierte entradas precisas (crisp inputs) en conjuntos difusos. Por ejemplo, convierte una temperatura de 25°C en grados de pertenencia a conjuntos como "baja", "media" y "alta".
2. **Base de Reglas**: Un conjunto de reglas if-then que describen cómo se deben tomar decisiones basadas en las entradas difusas. Por ejemplo:
   - "SI la temperatura es alta Y la humedad es alta, ENTONCES la velocidad del ventilador es rápida".
   - Las reglas están basadas en la experiencia o el conocimiento experto.
3. **Motor de Inferencia**: Evalúa las reglas en la base de reglas utilizando las entradas difusas y combina los resultados de las reglas para determinar la acción de salida difusa. Los métodos comunes incluyen el de Mamdani y el de Sugeno.
  - **En el método Mamdani**, esto generalmente se hace tomando el máximo de las salidas difusas superpuestas.

4. **Defuzzificación**: Convierte la salida difusa en una salida precisa (crisp output). Esto puede hacerse usando varios métodos, como el centro de gravedad o el método del promedio ponderado.

## Ejemplo de Funcionamiento

Imagina un controlador difuso para un sistema de aire acondicionado:

1. **Entradas**: Temperatura y Humedad.
   - Temperatura: 25°C
   - Humedad: 70%
2. **Fuzzificación**:
   - Temperatura de 25°C podría tener una pertenencia de 0.6 al conjunto "media" y 0.4 al conjunto "alta".
   - Humedad de 70% podría tener una pertenencia de 0.8 al conjunto "alta".
3. **Base de Reglas**:
   - Regla 1: "SI la temperatura es media Y la humedad es alta, ENTONCES la velocidad del ventilador es media".
   - Regla 2: "SI la temperatura es alta Y la humedad es alta, ENTONCES la velocidad del ventilador es rápida".
4. **Motor de Inferencia**:
   - Evaluación de las reglas usando las pertenencias difusas. Aquí se combinarían las pertenencias según las reglas para determinar la acción de salida.
   - Por ejemplo, una pertenencia combinada para la velocidad del ventilador "media" podría ser 0.6 y para "rápida" podría ser 0.4.
5. **Defuzzificación**:
   - Convierte estas salidas difusas en una única salida precisa. Si se usa el método del centro de gravedad, se calcularía un valor preciso para la velocidad del ventilador.

## Ventajas del Control Difuso

1. **Manejo de Incertidumbre**: Los controladores difusos son excelentes para sistemas con información imprecisa o incierta.
2. **No requiere un modelo exacto**: A diferencia de los controladores clásicos, no necesitan un modelo matemático exacto del sistema a controlar.
3. **Facilidad de Incorporar Experiencia Humana**: Las reglas pueden basarse en el conocimiento de expertos.

## Aplicaciones

- Control de temperatura
- Control de motores
- Sistemas de navegación autónoma
- Electrodomésticos inteligentes (lavadoras, refrigeradores)
- Sistemas de decisión en ambientes inciertos

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Definir los rangos de las variables de entrada y salida
temperature = ctrl.Antecedent(np.arange(0, 41, 1), 'temperature')
humidity = ctrl.Antecedent(np.arange(0, 101, 1), 'humidity')
fan_speed = ctrl.Consequent(np.arange(0, 101, 1), 'fan_speed')

# Definir las funciones de pertenencia para la temperatura
temperature['low'] = fuzz.trimf(temperature.universe, [0, 0, 20])
temperature['medium'] = fuzz.trimf(temperature.universe, [10, 25, 40])
temperature['high'] = fuzz.trimf(temperature.universe, [30, 40, 40])

# Definir las funciones de pertenencia para la humedad
humidity['low'] = fuzz.trimf(humidity.universe, [0, 0, 50])
humidity['medium'] = fuzz.trimf(humidity.universe, [30, 50, 70])
humidity['high'] = fuzz.trimf(humidity.universe, [60, 100, 100])

# Definir las funciones de pertenencia para la velocidad del ventilador
fan_speed['low'] = fuzz.trimf(fan_speed.universe, [0, 0, 50])
fan_speed['medium'] = fuzz.trimf(fan_speed.universe, [30, 50, 70])
fan_speed['high'] = fuzz.trimf(fan_speed.universe, [60, 100, 100])

# Definir las reglas difusas
rule1 = ctrl.Rule(temperature['low'] & humidity['low'], fan_speed['low'])
rule2 = ctrl.Rule(temperature['low'] & humidity['medium'], fan_speed['medium'])
rule3 = ctrl.Rule(temperature['low'] & humidity['high'], fan_speed['medium'])
rule4 = ctrl.Rule(temperature['medium'] & humidity['low'], fan_speed['medium'])
rule5 = ctrl.Rule(temperature['medium'] & humidity['medium'], fan_speed['high'])
rule6 = ctrl.Rule(temperature['medium'] & humidity['high'], fan_speed['high'])
rule7 = ctrl.Rule(temperature['high'] & humidity['low'], fan_speed['high'])
rule8 = ctrl.Rule(temperature['high'] & humidity['medium'], fan_speed['high'])
rule9 = ctrl.Rule(temperature['high'] & humidity['high'], fan_speed['high'])

# Crear el sistema de control difuso
fan_speed_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9])
fan_speed_simulation = ctrl.ControlSystemSimulation(fan_speed_ctrl)

# Establecer los valores de entrada
fan_speed_simulation.input['temperature'] = 30
fan_speed_simulation.input['humidity'] = 100

# Calcular el resultado
fan_speed_simulation.compute()

# Imprimir el resultado
print(f"Fan speed: {fan_speed_simulation.output['fan_speed']}")

# Visualizar el resultado
temperature.view(sim=fan_speed_simulation)
humidity.view(sim=fan_speed_simulation)
fan_speed.view(sim=fan_speed_simulation)
```

Diseño de un sistema que a través de redes Neuronales pueda reconocer figuras Geometricas planas

```python
#Recolección de Datos
#Podemos usar Python con bibliotecas como OpenCV para generar imágenes
#de figuras geométricas.

import cv2
import numpy as np
import os

# Crear carpetas para guardar las imágenes
os.makedirs('shapes/circle', exist_ok=True)
os.makedirs('shapes/square', exist_ok=True)
os.makedirs('shapes/triangle', exist_ok=True)
os.makedirs('shapes/rectangle', exist_ok=True)

# Función para generar imágenes
def create_shape(shape, path, num_images=1000):
    for i in range(num_images):
        image = np.zeros((64, 64, 3), dtype=np.uint8)
        if shape == 'circle':
            cv2.circle(image, (32, 32), 20, (255, 255, 255), -1)
        elif shape == 'square':
            cv2.rectangle(image, (16, 16, 48, 48), (255, 255, 255), -1)
        elif shape == 'triangle':
            points = np.array([[32, 8], [8, 56], [56, 56]], np.int32)
            cv2.fillPoly(image, [points], (255, 255, 255))
        elif shape == 'rectangle':
            cv2.rectangle(image, (8, 16, 56, 48), (255, 255, 255), -1)
        cv2.imwrite(f"{path}/{shape}_{i}.png", image)

# Generar las imágenes
create_shape('circle', 'shapes/circle')
create_shape('square', 'shapes/square')
create_shape('triangle', 'shapes/triangle')
create_shape('rectangle', 'shapes/rectangle')
```

```python
#2. Preprocesamiento de Datos
#Cargamos y preprocesamos las imágenes para el
#entrenamiento de la red neuronal
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Definir los generadores de datos
datagen = ImageDataGenerator(rescale=0.2, validation_split=0.2)

train_generator = datagen.flow_from_directory(
    'shapes/',
    target_size=(64, 64),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

validation_generator = datagen.flow_from_directory(
    'shapes/',
    target_size=(64, 64),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)
```

```python
#3. Diseño de la Red Neuronal
#Definimos una red neuronal convolucional (CNN)
#para la clasificación de las figuras geométricas
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Definir la arquitectura de la red neuronal
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(4, activation='softmax')  # 4 clases: círculo, cuadrado, triángulo, rectángulo
])

# Compilar el modelo
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
model.fit(train_generator, epochs=2, validation_data=validation_generator)
```

```python
#4. Evaluación del Modelo
#Evaluamos el rendimiento del modelo con el conjunto de datos de prueba.
loss, accuracy = model.evaluate(validation_generator)
print(f"Loss: {loss}, Accuracy: {accuracy}")
```

```python
#5. Implementación del Sistema
#Creamos una función que tome una imagen como entrada
#y devuelva la figura geométrica reconocida.

import matplotlib.pyplot as plt

# Diccionario para mapear índices a nombres de figuras
class_labels = {v: k for k, v in train_generator.class_indices.items()}

def recognize_shape(image_path):
    image = cv2.imread(image_path)
    image = cv2.resize(image, (64, 64))
    image = np.expand_dims(image, axis=0) / 255.0
    prediction = model.predict(image)
    class_idx = np.argmax(prediction)
    return class_labels[class_idx]

# Ejemplo de uso
image_path = 'shapes/circle/circle_0.png'
recognized_shape = recognize_shape(image_path)
print(f"La forma de la Imagen es: {recognized_shape}")

# Mostrar la imagen
image = cv2.imread(image_path)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title(f"Forma Reconocida: {recognized_shape}")
plt.show()
```

# Función de activación RELU

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de activación ReLU
def relu(x):
    return np.maximum(0, x)

# Generar datos para x
x = np.linspace(-10, 10, 400)

# Aplicar la función ReLU a los datos
y = relu(x)

# Crear el gráfico
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="ReLU(x)")
plt.title("Función de Activación ReLU", fontsize=16)
plt.xlabel("x", fontsize=14)
plt.ylabel("ReLU(x)", fontsize=14)
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend(fontsize=14)
plt.show()
```
